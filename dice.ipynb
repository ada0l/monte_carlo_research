```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
np.random.seed(1834)
N = 1000000
dice_1 = np.random.randint(1, 6 + 1, size=N)
dice_2 = np.random.randint(1, 6 + 1, size=N)
sum_dices = dice_1 + dice_2
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, sharey=True, figsize=(24,6))
sns.countplot(dice_1, ax=ax1)
sns.countplot(dice_2, ax=ax2)
sns.countplot(sum_dices, ax=ax3)
```
