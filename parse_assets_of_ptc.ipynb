```python
import requests
url = 'https://ru.wikipedia.org/wiki/%D0%98%D0%BD%D0%B4%D0%B5%D0%BA%D1%81_%D0%A0%D0%A2%D0%A1'
r = requests.get(url)
r
```

```python
from lxml import html
tree = html.fromstring(r.text)
table = tree.xpath('/html/body/div[3]/div[3]/div[5]/div[1]/table[5]/tbody')[0]
res = []
for item in table.getchildren()[1:]:
    res.append(item.xpath("td[2]")[0].text.strip())
```

```python
from lxml import html
tree = html.fromstring(r.text)
table = tree.xpath('/html/body/div[3]/div[3]/div[5]/div[1]/table[5]/tbody')[0]
res = dict()
for item in table.getchildren()[1:]:
    #res.append(item.xpath("td[2]")[0].text.strip())
    code, name = item.xpath("td[2]")[0].text.strip(), item.xpath("td[3]")[0].text.split(',')[0]
    res[code] = name
res
```

```python
import pandas as pd
df = pd.DataFrame(res, columns=["code", "name"])
df.to_csv("data/ptc_assets.csv")
df
```
