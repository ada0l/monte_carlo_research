```python
import pandas_datareader as web
import datetime

import pandas as pd
```

```python
assets = ['GAZP',
 'SBER',
 'SBERP',
 'LKOH',
 'GMKN',
 'YNDX',
 'NVTK',
 'TATN',
 'TATNP',
 'ROSN',
 'SNGS',
 'SNGSP',
 'MGNT',
 'FIVE',
 'MTSS',
 'POLY',
 'ALRS',
 'CHMF',
 'PLZL',
 'IRAO',
 'NLMK',
 'VTBR',
 'MOEX',
 'PHOR',
 'TRNFP',
 'MAGN',
 'RTKM',
 'RUAL',
 'AFLT',
 'PIKK',
 'HYDR',
 'FEES',
 'AFKS',
 'LSRG',
 'CBOM',
 'UPRO',
 'DSKY',
 'LNTA',
 'SFIN',
 'RNFT',
 'MVID']
date_end = datetime.datetime.today()
date_start = date_end - datetime.timedelta(days=5 * 365)
data = {asset: web.DataReader(asset, 'moex', start=date_start, end=date_end) for asset in assets}
```

```python
df_series = {asset: data[asset]["CLOSE"] for asset in assets}
```

```python
df = pd.DataFrame(df_series)
df.sort_index(inplace=True)
df
```

```python
df.to_csv("./data/portfolio_optimization_2.csv")
```
